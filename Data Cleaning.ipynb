### Modifying a DataFrame

```python
df_cleaned = df.drop(df[df['Score'] < 0].index)
df_cleaned
```

```python
df_cleaned = df_cleaned.drop('City', axis = 1)
df_cleaned
```

```python
df_mean = df_cleaned['Age'].mean()
df_cleaned['Age'] = (df_cleaned['Age'].fillna(df_mean))

df_cleaned
```

```python
score_mean = df_cleaned['Score'].mean()
df_cleaned['Score'] = df_cleaned['Score'].fillna(score_mean)
df_cleaned
```

```python
df_cleaned = df_cleaned.dropna()
df_cleaned
```
