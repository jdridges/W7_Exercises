```python
import numpy as np 
import pandas as pd
```

### Series

```python
aa = [1,2,3,4]
ab = (5,6,7,8)
ac = [9, 10, 11, 12]
ad = {'first' : 13, 'second' : 14, 'third':15, 'fourth':16}
```

```python
ser1 = pd.Series(aa)
print(ser1)
```

```python
ser2 = pd.Series(ab)
print(ser2)
```

```python
ser3 = pd.Series(ac)
```

```python
ser4 = pd.Series(ad)
print(ser4)
```

```python
ser5 = pd.Series(17,18,19,20)
print(ser5)

#typeerror: Index(...) must be called with a collection of some kind, 18 was passed
```

### DataFrames from Series

```python
df1 = pd.DataFrame(ser1)
df1
```

```python
df1.insert(1, 1, (ser2))
df1
```

```python
df1.insert(2, 2, (ser3))
df1
```

```python
df1.insert(3, 3, (ser4))
df1

#df1 returns NaN values from ser4
```

```python
df1.loc[4] = [5,9,12,0]
```

```python
df1
```
