```python
import numpy as np
import matplotlib.pyplot as plt
```

### Creating Basic Plots

```python
x = [1,2,3,4,5]
y = [2,3,5,7,11]
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x,y, marker='o')
plt.show()
```

```python
plt.title('My First Graph')
plt.xlabel('X Ray')
plt.ylabel('Y Ray')

plt.plot(x, y, marker='o')
plt.show()
```

### Styling Plots

```python
plt.title('My First Graph')
plt.xlabel('X Ray')
plt.ylabel('Y Ray')

plt.plot(x, y, marker='o', ls= '--', c='red')
plt.show()
```

```python
plt.title('My First Graph')
plt.xlabel('X Ray')
plt.ylabel('Y Ray')

plt.plot(x, y, marker='s', ls= '--', c='red', ms= 10, mfc='g')
plt.show()
```

```python
plt.title('My First Graph')
plt.xlabel('X Ray')
plt.ylabel('Y Ray')

plt.plot(x, y, marker='*', ls= '-.', c='m', ms= 30, mfc='c', )
plt.show()
```

```python
plt.title('My First Graph')
plt.xlabel('X Ray')
plt.ylabel('Y Ray')

plt.plot(x, y, marker='*', ls= '-.', c='m', ms= 100, mfc='c', )
plt.show()
```
