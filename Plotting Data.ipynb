```python
import numpy as np
import matplotlib.pyplot as plt
```

### Creating Basic Plots

```python
x = [1,2,3,4,5]
y = [2,3,5,7,11]
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x,y, marker='o')
plt.show()
```

```python
plt.title('My First Graph')
plt.xlabel('X Ray')
plt.ylabel('Y Ray')

plt.plot(x, y, marker='o')
plt.show()
```

### Styling Plots

```python
plt.title('My First Graph')
plt.xlabel('X Ray')
plt.ylabel('Y Ray')

plt.plot(x, y, marker='o', ls= '--', c='red')
plt.show()
```

```python
plt.title('My First Graph')
plt.xlabel('X Ray')
plt.ylabel('Y Ray')

plt.plot(x, y, marker='s', ls= '--', c='red', ms= 10, mfc='g')
plt.show()
```

```python
plt.title('My First Graph')
plt.xlabel('X Ray')
plt.ylabel('Y Ray')

plt.plot(x, y, marker='*', ls= '-.', c='m', ms= 30, mfc='c', )
plt.show()
```

```python
plt.title('My First Graph')
plt.xlabel('X Ray')
plt.ylabel('Y Ray')

plt.plot(x, y, marker='*', ls= '-.', c='m', ms= 100, mfc='c', )
plt.show()
```

### Advanced Plotting Techniques

```python
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [23,17,35,29]

plt.bar(products, sales)
plt.show()
```

```python
plt.title('Sales by Product')
plt.xlabel('Product')
plt.ylabel('Sales ($)')

plt.bar(products, sales, color = 'skyblue', width = .5)
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y)
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
s = np.random.rand(50) * 100
c = np.random.rand(50)
plt.scatter(x, y, s = s, c = c, cmap = 'viridis')
plt.show()
```

### Subplots

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 5))

x = np.linspace(0, 10, 100)
y = np.sin(x)
ax1.plot(x, y, color = 'blue')
ax1.set_title('Sine Wave')
ax1.set_xlabel('X-Axis')
ax1.set_ylabel('Y-Axis')

data = np.random.randn(1000)
ax2.hist(data)
ax2.set_title('Histogram')
ax2.set_xlabel('X-Axis')
ax2.set_ylabel('Y-Axis')

plt.show()
```
